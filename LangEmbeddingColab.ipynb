```python
from transformers import BertTokenizer

tokenizer = BertTokenizer("nih_vocab.txt", do_basic_tokenize=False)
print(tokenizer.get_vocab())
print(tokenizer("nibble groom"))
print(tokenizer.convert_tokens_to_ids(tokenizer.tokenize("nibble groom")))
```

```python
from transformers import LineByLineTextDataset

dataset = LineByLineTextDataset(
    tokenizer=tokenizer,
    file_path="all_30s_sequences.txt",
    block_size=256,
)
```

```python
from transformers import BertConfig, BertForMaskedLM
config = BertConfig(
    vocab_size=31,
    hidden_size=768, 
    num_hidden_layers=6, 
    num_attention_heads=12,
    max_position_embeddings=256
)
# config = BertConfig(vocab_size=len(tokenizer.get_vocab()), hidden_size=36, num_hidden_layers=4, model_max_length=100)
model = BertForMaskedLM(config)
```

```python
from transformers import DataCollatorForLanguageModeling
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm_probability=0.15)
```

```python
from transformers import Trainer, TrainingArguments

training_args = TrainingArguments(
    output_dir=".",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    num_train_epochs=10,
    weight_decay=0.01,
    remove_unused_columns=False
)

trainer = Trainer(
    model=model,
    args=training_args,
    tokenizer=tokenizer,
    train_dataset=dataset,
    eval_dataset=dataset,
    data_collator=data_collator
    )

trainer.train()
```

```python
trainer.train()
```
